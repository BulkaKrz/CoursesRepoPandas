```python
import pandas as pd
import numpy as np
fortune = pd.read_csv('.\course-files\Fortune_500_2017.csv', usecols=['Rank','Title','Employees','Profits','Assets'], index_col='Rank')
```

```python
fortune.head()
```

```python
fortune['Employees'].rank(ascending=False).head()
```

```python
fortune['RankByEmployee'] = fortune['Employees'].rank(ascending=False)
```

```python
fortune.head()
```

```python
fortune.sort_values(by='RankByEmployee').head()
```

```python
fortune['Profits'].rank(ascending=False).head()
```

```python
fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
fortune.head()
```

```python
fortune.nlargest(n=3,columns='Assets')
```

```python
fortune.nsmallest(n=3,columns='Assets')
```
