```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('.\course-files\Fortune_500_2017.csv', usecols=['Rank','Title','Industry','Hqlocation','Sector'])
```

```python
fortune.head()
```

```python
fortune['Sector'] = fortune['Sector'].str.upper()
```

```python
fortune.head()
```

```python
compSearch = fortune['Industry'].str.lower().str.contains('comp')
```

```python
fortune[compSearch].head()
```

```python
fortune[['City','State']] = fortune['Hqlocation'].str.split(', ', expand=True)
```

```python
fortune.head()
```

```python
fortune.axes
```

```python
def BuildShortcut(row):
    industry = row['Industry']
    result = ''
    for i in industry.split(' '):
        result += i[0]
    return result
```

```python
BuildShortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(BuildShortcut, axis=1)
```
