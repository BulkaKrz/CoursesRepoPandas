```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('./course-files/fuel.csv', usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'])
fuel.head()
```

```python
fuel.drop_duplicates(subset=['Year','Make','Model'],inplace=True)
```

```python
fuel.dropna(how='any',inplace=True)
```

```python
fuel.set_index(['Year','Make','Model'], inplace=True)
```

```python
fuel.head()
```

```python
fuel.stack().head()
```

```python
fuel.stack().to_frame().head()
```

```python
fuel.unstack().head()
```

```python
fuel.unstack(level=0)
```

```python
fuel.stack().unstack(level=0)
```

```python
fuel.stack().unstack(level='Make')
```

```python
fuel.stack().unstack(level=['Year','Make'])
```

```python
fuel.stack().unstack(level=['Make','Year']).sort_index(axis=1)
```
