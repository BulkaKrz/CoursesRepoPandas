```python
import pandas as pd
```

```python
import numpy as np
```

```python
animals = pd.read_csv('.\course-files\sleep_time.csv', index_col='ID').fillna(0)
```

```python
animals.head()
```

```python
herbi = animals.where(animals.vore == 'herbi')
```

```python
len(herbi)
```

```python
herbi
```

```python
len(herbi.dropna())
```

```python
carni = animals.where(animals.vore == 'carni')
```

```python
print(len(carni))
print(len(carni.dropna()))
```

```python
omni = animals.where(animals.vore == 'omni')
print(len(omni))
print(len(omni.dropna()))
```

```python
print('herbi sleep - ',herbi['sleep_total'].mean())
print('carni sleep - ',carni['sleep_total'].mean())
print('omni sleep - ',omni['sleep_total'].mean())
```

```python
animals.query('sleep_total > 14 and vore == "herbi"')
```

```python
animals.query('sleep_total > 14 and vore == "herbi" and bodywt > 1')
```
