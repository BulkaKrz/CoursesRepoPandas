```python
import pandas as pd
import numpy as np
```

```python
url = 'https://ec.europa.eu/eurostat/estat-navtree-portlet-prod/BulkDownloadListing?file=data/isoc_cimobi_frq.tsv.gz&unzip=true'
```

```python
df = pd.read_csv(url, delimiter='\t')
```

```python
df.head()
```

```python
df_desc = df['indic_is,ind_type,unit,time\\geo'].str.split(',', expand=True)
```

```python
df_desc.head()
```

```python
df_desc.rename({0: 'indic_is', 1: 'ind_type', 2:'unit', 3:'time\\geo'}, inplace=True, axis=1)
```

```python
df_desc.head()
```

```python
df_desc = df_desc.join(df)
```

```python
df_desc.drop(labels='indic_is,ind_type,unit,time\\geo', axis=1, inplace = True)
```
