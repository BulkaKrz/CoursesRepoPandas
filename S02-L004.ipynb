```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
import random as rmd
```

```python
 dataAsFloatList = []
```

```python
for i in range(100000):
    dataAsFloatList.append(i*rmd.random())
```

```python
dataAsFloatList
```

```python
dataAsFloatSeries = pd.Series(dataAsFloatList)
```

```python
dataAsFloatSeries
```

```python
print('size \t {}'.format(dataAsFloatSeries.size))
print('nbytes \t {}'.format(dataAsFloatSeries.nbytes))
print('shape \t {}'.format(dataAsFloatSeries.shape))
print('axes \t {}'.format(dataAsFloatSeries.axes))
print('dtype \t {}'.format(dataAsFloatSeries.dtype))
print('index \t {}'.format(dataAsFloatSeries.index))
print('is_unique \t {}'.format(dataAsFloatSeries.is_unique))
print('is_monotonic \t {}'.format(dataAsFloatSeries.is_monotonic))
```

```python
dataAsStringList =[]
```

```python
for i in range(10000):
    dataAsStringList.append(str(i*rmd.random()))
```

```python
dataAsStringSeries = pd.Series(dataAsStringList)
```

```python
print('size \t {}'.format(dataAsStringSeries.size))
print('nbytes \t {}'.format(dataAsStringSeries.nbytes))
print('dtype \t {}'.format(dataAsStringSeries.dtype))
```
